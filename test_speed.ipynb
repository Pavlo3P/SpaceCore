# SpaceCore Overhead Diagnostics

This notebook is a thin entry point. The benchmark source of truth is the JSON artifact written by `bench/run.py`; the HTML view is rendered from that JSON.

````python
from pathlib import Path
import subprocess
import sys
from IPython.display import HTML, display

out_json = Path('bench/out/overhead.json')
out_html = Path('bench/out/overhead.html')
subprocess.run([
    sys.executable,
    'bench/run.py',
    '--json', str(out_json),
    '--html', str(out_html),
], check=True)

display(HTML(out_html.read_text()))
````

For a broader anomaly scan, run:

```bash
python bench/run.py --suite full --json bench/out/overhead-full.json --html bench/out/overhead-full.html
```
